```python

import pandas as pd

def main():
  # load the data set
  dev_df =  pd.read_csv('https://raw.githubusercontent.com/QUANHONGLE/CS421-emotion-prediction/main/P1_data/trac2_CONVT_dev.csv', on_bad_lines='skip')
  # getting the conversation counts for the ids
  conversation_counts = dev_df.groupby('conversation_id')['turn_id'].count()
  # filtering out the one's with atleast 10 utterences
  filtered_conversations = conversation_counts[conversation_counts >= 10]
  # getting the 5 conversation id's to be used
  five_conversation_ids = filtered_conversations.head(5).index.tolist()
  print(five_conversation_ids)
  # only take the conversations that are in the 5 id's
  convo = dev_df[dev_df['conversation_id'].isin(five_conversation_ids)]
  convo.to_csv("5_convos.csv", index=False)
  # extract the conversations and id's for llm analysis and convert to csv
  conv_text_to_llm = convo[['conversation_id', 'text']]
  print(conv_text_to_llm)
  conv_text_to_llm.to_csv("extracted_texts.csv", index=False)

main()
```
